```python
import pandas as pd

kc_house_data = pd.read_csv('../linear_regression/datasets/kc_house_data.csv')

#leaving only necessary columns in the dataframe
new_kc_house_data = kc_house_data.loc[:, ~kc_house_data.columns.isin(['id','data','zipcode','lat','long',])]
```

```python
import seaborn as sb
import matplotlib.pyplot as plt

#creating a figure to show an heatmap
plt.figure(figsize=(30,30))

#creating the heatmap
sb.heatmap(new_kc_house_data.corr())

#showing the heatmap
plt.show()
```
